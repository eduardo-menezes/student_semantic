```python
import pandas as pd
import numpy as np
import json
import seaborn as sns
import matplotlib.pyplot as plt
```

# Reading data into a python dict

```python
with open("data/raw/students/students.json", "r") as f:
    students_dict = json.load(f)
```

```python
with open("data/raw/missed_days/missed_days.json", "r") as f:
    missed_days_dict = json.load(f)
```

# Transforming dict keys in columns and dict values in tables records

```python
missed_days = pd.json_normalize(missed_days_dict['missed_classes'], sep='_')
missed_days
```

```python
students = pd.json_normalize(students_dict['students'], sep='_')
students
```

# Verifying duplicates

```python
students.student_id.value_counts()
```

# Build OBT

```python
students_grades_and_missed_days = students.merge(missed_days, how='left', on='student_id')
```

```python
students_grades_and_missed_days.describe()
```

## Question 2: What's the overall avarage per student?

```python
students_grades_and_missed_days['overall_mean'] = students_grades_and_missed_days[['grades_math','grades_science','grades_history','grades_english']].mean(axis=1)
students_grades_and_missed_days
```

```python
min_overall_mean = students_grades_and_missed_days['overall_mean'].min()
max_overall_mean = students_grades_and_missed_days['overall_mean'].max()
```

```python
students_grades_and_missed_days['overall_mean'].plot(kind='bar')
print(
f"""
overall_mean seems to follow a uniform distribution
with min value as {min_overall_mean} and max value as {max_overall_mean}

"""
)
```

## Question 3. What percent of students have math grades higher than 90?

```python
students_math_higher_90 = (students_grades_and_missed_days['grades_math'] > 90).sum()
percet_higher_90 = students_math_higher_90/len(students_grades_and_missed_days['grades_math'])
percet_higher_90

print(f"""{percet_higher_90 *100}% have 90 or higher in math""")
```

## Question 3. What percent of students have math grades less than 80?

```python
students_math_less_80 = (students_grades_and_missed_days['grades_math'] <= 80).sum()
percet_math_less_80 = students_math_less_80/len(students_grades_and_missed_days['grades_math'])
percet_math_less_80

print(f"""{percet_math_less_80 *100}% have 80 or less in math""")
```
